Introducción:
 Crear un modelo que hará la predicción de la calificación de un grupo de estudiantes tomando en cuenta el número de horas dedicadas a estudiar.

Para iniciar, es fundamental, al trabajar en este tipo de proyectos, cargar las bibliotecas necesarias.

```python
# Importar librerías
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Configurar la semilla aleatoria para reproducibilidad
np.random.seed(0)

# Generar datos
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2

# Dividir datos en conjunto de entrenamiento y conjunto de prueba (80% train, 20% test)
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2)

# Entrenar el modelo de regresión lineal
regression_model = LinearRegression()
regression_model.fit(train_data, train_labels)

# Realizar predicciones con el conjunto de prueba
predictions = regression_model.predict(test_data)

# Calcular el error cuadrático medio
mse = mean_squared_error(test_labels, predictions)
print("Error Cuadrático Medio:", mse)

# Graficar el modelo de regresión lineal
plt.scatter(test_data, test_labels, color='black')
plt.plot(test_data, predictions, color='blue', linewidth=3)
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones')
plt.show()
```
